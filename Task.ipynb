```python
import pandas as pd

pokemon = pd.read_csv("https://gist.githubusercontent.com/simsketch/1a029a8d7fca1e4c142cbfd043a68f19/raw/bd584ee6c307cc9fab5ba38916e98a85de9c2ba7/pokemon.csv", error_bad_lines=False)
```

```python
pokemon.head(5)
```

```python
pokemon.isnull().sum()
```

```python
#We then remove redundant and hard to deal with features.
features_to_remove = ['TYPE2', 'NUMBER', 'CODE', 'SERIAL', 'COLOR', 'ABILITY1', 'ABILITY2','ABILITY HIDDEN', 'MEGA_EVOLUTION', 'HEIGHT', 'WEIGHT', 'HP', 'SPD', 'TOTAL']

pokemon = pokemon.drop(features_to_remove, axis = 1)
```

```python
pokemon.head(5)
```

```python
pokemon.tail(5)
```

```python
pokemon_stats = (pokemon['ATK'] - pokemon['DEF'])*0.8 + (pokemon['SP_ATK'] - pokemon['SP_DEF'])*0.2
```

```python
pokemon_stats
```

```python
pokemon_new = pd.concat([pokemon, pokemon_stats], axis=1)
```

```python
pokemon_new
```

```python
pokemon_new.columns = ['NAME','TYPE1', 'GENERATION', 'LEGENDARY', 'ATK', 'DEF','SP_ATK', 'SP_DEF', 'POKEMON_STATS']
pokemon_new
```

```python
len(pokemon_new['NAME'].unique().tolist())
```

```python
pokemon_strength = pokemon_new[['NAME','GENERATION', 'TYPE1','LEGENDARY','POKEMON_STATS']]
```

```python
pokemon_strength.sort_values(by='POKEMON_STATS',ascending=True).iloc[:20,:]
```
